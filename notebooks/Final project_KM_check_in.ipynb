```python
# This tells matplotlib not to try opening a new window for each plot.
%matplotlib inline

# General libraries.
import re
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# SK-learn libraries for learning.
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.grid_search import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# SK-learn libraries for evaluation.
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import classification_report

# The competition datafiles are in the directory ../input
# Read competition data files:
train = pd.read_csv("train.csv")
test  = pd.read_csv("test.csv")

train.head()
```

```python
train.head()
```

```python
train_data = train.drop('Id', axis=1)
test_data = test.drop('Id', axis=1)

# print(test_data.shape)
# print(train_data.shape)

train_labels = train_data['Cover_Type']
train_data = train_data.drop('Cover_Type',axis=1)

print(test_data.shape)
print(train_data.shape)
```

```python
#Logistic Regression baseline
#try logistic regression
logit = LogisticRegression()
logit.fit(train_data, train_labels)
test_data['Cover_Type'] = pd.Series(logit.predict(test_data))



logit_submission = pd.concat([test['Id'].astype(int), test_data['Cover_Type']], axis=1)
logit_submission = logit_submission.set_index('Id')
logit_submission.to_csv('logit_submission.csv')

print(test_data.shape)
#print(submission.shape)

#My logistic regression baseline submission scored 0.55306 according to kaggle
```

```python
#Random Forests baseline
#Read competition data files:
train = pd.read_csv("train.csv")
test  = pd.read_csv("test.csv")

#Remove the first column
train_data = train.drop('Id', axis=1)
test_data = test.drop('Id', axis=1)

train_labels = train_data['Cover_Type']
train_data = train_data.drop('Cover_Type',axis=1)

print(test_data.shape)
print(train_data.shape)


rndfs= RandomForestClassifier()
rndfs.fit(train_data,train_labels)
test_data['Cover_Type'] = pd.Series(rndfs.predict(test_data))
rndfs_submission = pd.concat([test['Id'].astype(int), test_data['Cover_Type']], axis=1)
rndfs_submission = rndfs_submission.set_index('Id')
rndfs_submission.to_csv('rndfs_submission.csv')

#Random forests Your submission scored 0.71391
```
